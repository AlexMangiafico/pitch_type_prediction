```python
from pybaseball import statcast
import pandas as pd
import contextlib
```

```python
start_date = "2024-01-01"
end_date = "2024-12-31"
with contextlib.redirect_stdout(None):
    for date in pd.date_range(start=start_date,end=end_date):
        day_pitches = statcast(start_dt = date.strftime('%Y-%m-%d'))
        pitches = day_pitches if date.strftime('%Y-%m-%d') == start_date else pd.concat([pitches, day_pitches])
    pitches.to_csv('my_pitches.csv', index=False)
```
